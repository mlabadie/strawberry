```python
extract_at_module_scale
```

```python
analysis.crowntype_distribution(data=extract_at_module_scale,varieties=["Belle_et_Bonne"],crown_type="branch_crown")
```

```python
Mean= extract_at_module_scale.groupby(["Genotype", "order"]).mean()
sd= extract_at_module_scale.groupby(["Genotype", "order"]).std()

Mean
```

```python
analysis.pointwisemean_plot(data_mean=Mean,
                   data_sd=sd,
                   varieties=["Belle_et_Bonne"], 
                   variable='no_fruits',
                   title= "Pointwisemean of no_fruits",
                   ylab="Mean no_fruits",
                   expand= 0.1)
```

# Analysis

```python
from openalea.strawberry import analysis

meta_MTG_at_plant_scale=analysis.extract_at_plant_scale(meta_MTG)
meta_MTG_at_module_scale=analysis.extract_at_module_scale(meta_MTG)
meta_MTG_at_node_scale=analysis.extract_at_node_scale(meta_MTG)


pd.options.display.max_columns = None
pd.options.display.max_rows = None
meta_MTG_at_module_scale
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None

Mean= meta_MTG_at_module_scale.groupby(["Genotype", "order"]).mean()
sd= meta_MTG_at_module_scale.groupby(["Genotype", "order"]).std()
sd=sd.fillna(0)
Mean
```

```python
# genotypes=['Belle_et_Bonne','Belrubi','CF10122','CF10277','CF12074','CF9416','Ciflorette',
#            'Cigaline','Cireine','CIR107','Clery','Darselect','Dream','gariguette','Leroy Ladurie',
#            'Magnum','Manon','Moutot','Osiris','Rubis','Sanié','Sans rivale']

genotypes=['CF10277','CF12074','gariguette','CIR107']

analysis.pointwisemean_plot(data_mean=Mean,
                   data_sd=sd,
                   varieties=genotypes, 
                   variable='nb_total_leaves',
                   title= "Pointwisemean of nb_total_leaves",
                   ylab="Mean nb_total_leaves",
                   expand= 0.1)

```

```python
analysis.crowntype_distribution(data=meta_MTG_at_module_scale,varieties=genotypes,crown_type="branch_crown")
```
